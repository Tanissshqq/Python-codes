**lambda**

```python
def add(a,b):
    return a+b
add(10,5)
```

```python
add = lambda a,b : a+b
add(10,20)
```

```python
is_even = lambda a : a%2 == 0
is_even(8)
```

```python
greater = lambda a,b : a if  a>b else b
greater(5,7)
```
