**TRY AND EXCEPT**

```python
a = int(input('num 1 '))
b = int(input('num 2 '))
print(a/b)
```

```python
try:
    a = int(input('num 1 '))
    b = int(input('num 2 '))
    print(a/b)
except (ZeroDivisionError):
    print('num cannot divide by 0')
except (ValueError):
    print('enter value only')
finally:
    print('program end succesfully')
```

```python
try:
    a=int(input('num 1'))
except (ValueError):
    print('Taking default number')
    a=0
try:
    b=int(input('num 1'))
except (ValueError):
    print('Taking default number')
    b=0
try:
    print('div' ,a/b)
except (ZeroDivisionError):
    print('Value cannot be divided by zero')
finally:
    print('sum' ,a+b)
    print('mul' ,a*b)
```
