```python
class BankAccount:
    def __init__(self, name, username, password, balance=0):
        self.name = name
        self.username = username
        self.password = password
        self.balance = balance
        self.transaction = []
        self.transaction.append(f"Name: {name}")
        self.transaction.append(f"Opening balance: {balance}")
        print("WELCOME TO BANK")

    def login(self, username, password):
        if self.username == username and self.password == password:
            print("LOGIN SUCCESSFUL")
            return True
        else:
            print("INVALID USERNAME OR PASSWORD")
            return False

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transaction.append(
                f"Deposit amount is {amount} and balance is {self.balance}"
            )
        else:
            print("AMOUNT MUST BE POSITIVE")

    def viewbalance(self):
        print(f"Account Holder: {self.name}")
        print(f"Balance: {self.balance}")

    def show_transactions(self):
        for i in self.transaction:
            print(i)
```

```python
obj = BankAccount("Tanishq", "tanishq123", "pass123", 100)
```

```python
if obj.login("tanishq123", "pass123"):
    obj.deposit(500)
    obj.viewbalance()
    obj.show_transactions()
```
